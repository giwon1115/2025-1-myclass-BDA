## 시리즈활용

```python
import numpy as np
import pandas as pd

s=['희빈','수영','현수','지호','지민']
p=pd.Series([99,84,92,65,76],index=s)
p
```

```python
p.name='수학성적'
p.index.name='이름'
p
```

```python
p.sort_index() # ㅅ ㅇ ㅈ ㅎ 순서
```

```python
p.sort_index(ascending=False)
```

```python
p.sort_values(ascending=False)
```

##시리즈 기본 통계함수

```python
print(p.max())
print(p.min())
print(p.mean())
print(p.median())
print(p.count())
print(p.size)
```
