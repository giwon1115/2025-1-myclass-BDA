```python
df_word=pd.DataFrame({'word':nouns})
df_word.sample(5)
```

```python
df_word['len']=df_word['word'].str.len()
df_word=df_word.query('len>=2')
df_word.sample(5)
```

```python
df_word.query('word=="소년단"')
```

```python
df_word=df_word.groupby('word',as_index=False).agg(n=('word','count')).sort_values('n',ascending=False)
df_word
```

```python
from IPython.display import display_html
def display_side_by_side(*args):
    """여러 데이터프레임 비교가 쉽게 옆쪽으로 표시한다"""
    html_str=''
    for df in args:
        html_str += df.to_html() + '&nbsp;'*4
    display_html(html_str.replace('table','table style="display:inline"'), raw=True)
```

```python
display_side_by_side(df,df.explode(['A']))#시험예상
```

```python
display_side_by_side(df,df.explode(['B']))#시험예상
```

```python
display_side_by_side(df,df.explode(['A','B']))#시험예상
```
