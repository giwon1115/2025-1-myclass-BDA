##데이터프레임 수정

```python
df[['사망(명)','부상(명)']].sum(axis=1)
```

```python
df.iloc[:,1:].sum(axis=1)
```

```python
df2=df.copy()
df2['재해자수']=df[['사망(명)','부상(명)']].sum(axis=1)
df2 #새로운 열 추가
```

```python
df2.loc['총계']=df2.sum()
df2 #새로운 행 추가
```

```python
df2
```

```python
df2>10 #크면 True 작으면 False
```

```python
df[df2>10]
```
