#mpg데이터 활용 groupby() agg()

```python
mpg=pd.read_csv('data/mpg.csv')
mpg.info()
```

```python
mpg.sample(frac=.02)
```

```python
# 제조 회사 및 구동 방식별 분리
# cty 평균 구하기
mpg.groupby(['manufacturer', 'drv']).agg(mean_cty = ('cty', 'mean')) # 시험 예상
```

```python
mpg.groupby(['drv', 'manufacturer']).agg(mean_cty = ('cty', 'mean'))
```

```python
# 제조 회사와 차종별 분리
# cty 평균 구하기
# hwy 평균 구하기
mpg.groupby(['manufacturer', 'category']).agg(mean_cty = ('cty', 'mean'), mean_hwy = ('hwy', 'mean'))
```

```python
## audi의 drv별 빈도

# audi 추출
# drv별 분리
# 빈도 구하기
mpg.query('manufacturer == "audi"').groupby(['drv']).agg(n = ('drv', 'count'))
```

#merge시험예상문제

```python
import pandas as pd

df1 = pd.DataFrame([[1, 'apple'], [2, 'banana']],
                  columns = ['id', 'fruits'])
df1
```

```python
df2 = pd.DataFrame([[1, 3000], [5, 500]],
                  columns = ['id', 'price'])
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on='id',how='inner')
```

```python
pd.merge(df1,df2,on='id',how='left')
```

```python
df1.merge(df2,on='id',how='left')
```

```python
pd.merge(df1,df2,on='id',how='right')
```

```python
df1.merge(df2,on='id',how='right')
```

```python
df1.merge(df2, how='cross') # 두 프레임 모두에서 데카르트 곱을 생성하고 왼쪽 키의 순서를 유지
```
