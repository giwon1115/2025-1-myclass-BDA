## 시리즈활용

##시리즈 기본 통계함수

```python
print(p.max())
print(p.min())
print(p.mean())
print(p.median())
print(p.count())
print(p.size)
```

```python
s=pd.Series([3,1,1,2,None,np.nan,4,6,6,np.nan])
s
```

```python
s.values
```

```python
s.count()
```

```python
s.value_counts()
```

```python
s.value_counts(dropna=False)#가장 많은게 먼저 nan값도 포함하여 빈도계산
```

```python
s.value_counts(dropna=False,sort=False)# sort=False=시리즈에 등장한 순서대로(값 기준 X, 개수 기준 X) 결과를 보여줌
```

```python
s.value_counts(bins=4)
```

```python
s
```

```python
s.value_counts(bins=4,sort=False)
```

```python
s.value_counts(bins=[0.5,3.5,6.5],sort=False)
```
