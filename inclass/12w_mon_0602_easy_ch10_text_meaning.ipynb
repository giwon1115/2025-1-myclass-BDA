```python
FONT_PATH="/user/share/fonts/truetype/nanum/NanumGothic.ttf"
```

```python

from konlpy.tag import Hannanum
han = Hannanum()
han.nouns(u'대한민국의 영토는 한반도와 그 부속도서로 한다')
```

```python

nouns = han.nouns(moon)
type(nouns)
```

```python

len(nouns)
```

```python

nouns[:10]
```

```python
nouns[-10:]
```

```python
import pandas as pd
df_word = pd.DataFrame({'word': nouns})
df_word
```

```python

df_word[df_word.word == '대통령']
```

```python

df_word['len'] = df_word['word'].str.len()
df_word.head()
```

```python
df_word.query('len >= 2')
```

```python
df_word.query('len >= 1')
```

```python
df_word.query('len >= 2')
df_word.sort_values('len')
```

```python
df_word.query('word=="대통령"')
```

```python
df_freq=df_word.groupby('word',as_index=False)\
              .agg(n=('word','count'))\
              .sort_values('n',ascending=False)
df_freq
```

```python
df_freq.query('word=="대통령"')
```

```python

top20 = df_freq.head(20)
top20
```
